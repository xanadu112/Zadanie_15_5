```python
import pandas as pd
import matplotlib.pyplot as plt

prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]

df = pd.DataFrame(prices, columns=['Month', 'PricePLN'])

df.set_index('Month', inplace=True)

df['PriceUSD'] = df['PricePLN'] / 4

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['PriceUSD'], 'r--', label='Price in USD')
plt.xlabel('Month')
plt.ylabel('Price (USD)')
plt.title('Monthly Product Price in USD')
plt.legend()
plt.grid(True)
plt.show()
```
